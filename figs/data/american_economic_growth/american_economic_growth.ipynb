```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df = pd.read_csv("original_fig16-5.csv")
df.head()
```

```python
plt.rcParams['figure.figsize'] = [7.48, 3]
plt.rcParams['figure.dpi'] = 500
plt.rcParams['figure.facecolor'] = 'white'
plt.rcParams['axes.spines.right'] = False
plt.rcParams['axes.spines.top'] = False

BLUE = "#152B52"
GOLD = "#966B0F"
```

```python
colors = [BLUE if x < 1 else GOLD for x in df['Growth']]

plt.bar(x=df['Decade'], height=df['Growth'], color=colors, width=8)
plt.xticks(df['Decade'])
plt.xticks(rotation = 0)
plt.xlabel("")

plt.text(1950, 3.8, "Golden Age", weight='bold', ha='center', color=GOLD)
plt.plot([1926, 1974], [3.6, 3.6], lw=2, marker='|', color=GOLD)

plt.text(1995, 1.2, "Digital Stagnation", weight='bold', ha='center', color=BLUE)
plt.plot([1976, 2014], [1, 1], lw=2, marker='|', color=BLUE)

plt.ylabel("Percent\nper year", rotation=0, labelpad=35, weight='bold')

plt.tight_layout()
plt.savefig("golden.png")
```
